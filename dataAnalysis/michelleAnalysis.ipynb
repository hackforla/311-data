# Fuzzy Matching
Try to match each `RequestType` for the service requests to one or more expense through `ACCOUNT NAME`, `FUND NAME`, or `DEPARTMENT NAME`.

```python
request_types = request_df['RequestType'].unique()
expense_names = set(list(expense_df['ACCOUNT NAME'].unique()) +
                 list(expense_df['FUND NAME'].unique()) +
                 list(expense_df['DEPARTMENT NAME'].unique()))
```

```python
for request_type in request_types:
    print('Most similar expenses for {}'.format(request_type))
    display(process.extract(request_type, expense_names, limit=50))
```
